# Example 1
Find the root of exp(-x) - x = 0 in (0,1) 

```python
import numpy as np
from matplotlib import pyplot as plt

def fun(x):
    return np.exp(-x) - x

x = np.arange(0,1,0.01)
plt.plot(x,fun(x),x,0*x)
plt.savefig("Example 1 figure 1.jpg")
```

```python
# define the function and other parameters
def gfun(x):
    return np.exp(-x)

xinit = 1.0  # initial guess
epstol = 1.00/100
Nmax = 100

tracer=[[xinit,gfun(xinit)]]

xold = xinit
# iterations 
i=0
while (i<Nmax):
    x = gfun(xold)
    relError = np.abs((x-xold)/x)
    tracer.append([x,x])
    tracer.append([x,gfun(x)])
    if relError < epstol:
        print('The solution is found.')
        print('Iteration {}, x = {:.4f}, error = {:.3f}%'.format(i, x, relError*100))
        break;
    
    xold=x
    print('Iteration {}, x = {:.4f}, error = {:.3f}%'.format(i, x, relError*100))
    i=i+1
```

```python
tracer
```

```python
tracer=np.array(tracer)
x = np.arange(0,1.0,0.01)
plt.plot(x,x,x,gfun(x),tracer[:,0],tracer[:,1])
plt.savefig("Example 1 figure 2.jpg")
```

# Example 2
Find the root of x^2 + x -1 = 0 in (0,1) 

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt

def fun(x):
    return x**2 + x -1

x = np.arange(-2,1,0.01)
plt.plot(x,fun(x),x,0*x)
plt.savefig("Example 2 figure 1.jpg")
```

```python
xinit= 1.55  # initial guess
xold = xinit
epstol = 0.001
Nmax = 100

# define the function and other parameters
def gfun(x):
    #return 1/(1+x)
    #return 1 - x**2
    return (1-x)**0.5

# iterations 
i=0
tracer=[[xinit,gfun(xinit)]]
while (i<Nmax):
    x = gfun(xold)
    relError = np.abs((x-xold)/(x-xinit))
    tracer.append([x,x])
    tracer.append([x,gfun(x)])
    if relError < epstol:
        print('\nThe solution is found.')
        print('Iteration {}, x = {:.4f}, error = {:.5f}'.format(i, x, relError))
        break;
    
    xold=x
    print('Iteration {}, x = {:.4f}, error = {:.5f}'.format(i, x, relError))
    i=i+1
    
```

```python
tracer=np.array(tracer)
x = np.arange(0,2,0.01)
plt.plot(x,x,x,gfun(x),tracer[:,0],tracer[:,1])
plt.savefig("Example 2 figure 2.jpg")

```

## Example 3 
Find a solution of x^3 - x - 1 =0 in (1,2)

```python
import numpy as np
import sys

# define the function and its derivative
def gcs(x):
    return np.power(x+1, 1.0/3.0)

def dgcs(x):
    return np.power(x+1, -2.0/3.0)/3.0

xinit =1
xold= xinit  # initial guess
epstol = 0.001
Nmax = 100

# check convergence at the initial guess
if np.abs(dgcs(xold)) > 1:
    sys.exit('The convergence criterion, {:.3f}>1, is not satisfied at the initial guess.'.format(dgcs(xold)));
    
print('The convergence criterion, {:.3f}<1, is satisfied at the initial guess.'.format(dgcs(xold)))
tracer=[[xinit,gcs(xinit)]] 
# iteratons 
i=0
while (i<Nmax):
    x = gcs(xold)
    relError = np.abs((x-xold)/(x-xinit))
    if relError < epstol:
        print('\nA solution is found.')
        print('Iteration {}, x = {:.4f}, error = {:.5f}'.format(i, x, relError))
        tracer.append([x,x])
        tracer.append([x,gcs(x)])
        break
    
    xold=x
    print('Iteration {}, x = {:.4f}, error = {:.5f}'.format(i, x, relError))
    i=i+1
    
```

```python
tracer
```

```python
tracer=np.array(tracer)
x = np.arange(0,2,0.01)
plt.plot(x,x,x,gcs(x),tracer[:,0],tracer[:,1])

```
