### Example 1

```python
# Example 1

def fcs(x):
    return x*x*x-x-1

def dfcs(x):
    return 3*x*x-1
```

```python
import numpy as np
from matplotlib import pyplot as plt
#import matplotlib.pyplot as plt # either one is OK

# evenly sampled x
x = np.arange(0.1, 3, 0.01)
plt.xlim(0.1, 2.25)
plt.ylim(-4, 8)
# red dashes and green dashes
plt.plot(x, fcs(x), 'r', x, 0/x, 'g')
plt.xlabel('x', size=20)
plt.ylabel('f(x)', size=20)
plt.show()
```

```python
xinit= 2 # initial guess
xold = xinit
epstol =1e-16
Nmax = 100

# iterations
i=0
while (i<Nmax):
    i=i+1
    x = xold-fcs(xold)/dfcs(xold)
    relError = np.abs((x-xold)/(x-xinit))
    if relError < epstol:
        print('The solution is found.')
        print('Iteration {}, x = {:.4f}, error = {:.5e}'.format(i, x, relError))
        break;
    
    xold=x
    print('Iteration {}, x = {:.4f}, error = {:.5e}'.format(i, x, relError))
```

## Write a function for Newton-Raphson iteration for ease of use

```python
import numpy as np
from matplotlib import pyplot as plt

def newton_iteration(fun,dfun,xinit,epstol=1e-16,Nmax=100):
    i=0
    xold = xinit
    while (i<Nmax):
        i=i+1
        x = xold-fun(xold)/dfun(xold)
        err = np.abs((x-xold)/(x-xinit))
        if err < epstol:
            print('A solution is found.')
            print('Iteration {}, x = {:.4f}, error = {:.5e}'.format(i, x, err))
            break;
        xold=x
        print('Iteration {}, x = {:.4f}, error = {:.5e}'.format(i, x, err))

```

```python
def fcs(x):
    return x*x*x - 3*x*x + 2*x - 0.375

def dfcs(x):
    return 3*x*x-6*x+2

x = np.arange(0, 3.01, 0.01)
plt.xlim(0, 3)
#plt.ylim(-4, 8)
# red dashes and green dashes
plt.plot(x, fcs(x), 'r', x, 0*x, 'g')
plt.xlabel('x', size=20)
plt.ylabel('f(x)', size=20)
plt.show()
```

```python
x = np.arange(0, 3.01, 0.01)
plt.xlim(0, 2.25)
plt.ylim(-0.4, 0.1)
# red dashes and green dashes
plt.plot(x, fcs(x), 'r', x, 0*x, 'g')
plt.xlabel('x', size=20)
plt.ylabel('f(x)', size=20)
```

```python
xinit = 1.57735 # initial guess
xold = xinit
epstol = 1e-2
Nmax = 100

newton_iteration(fcs,dfcs,xinit,epstol,Nmax)
```

```python
xinit= 2 # initial guess
xold = xinit
epstol =1e-16
Nmax = 100

newton_iteration(fcs2,dfcs2,xinit,epstol)
```
